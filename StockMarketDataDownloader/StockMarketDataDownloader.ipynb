Using bsedata library
 
Documentation here : https://bsedata.readthedocs.io/en/latest/

Now get the stock Quotes.
This will return all the publicly available data, you can pick the fields you require from the dictionary

```python
from bsedata.bse import BSE
b= BSE()
b.getQuote('500010') # 500010.BO is the script code
```

Getting quotes in Bulk

```python
b = BSE()
codelist = ["500116", "512573"]
for code in codelist:
    quote = b.getQuote(code)
    print(quote["companyName"])
    print(quote["currentValue"])
    print(quote["updatedOn"])
```

Getting Top Gainers

```python
tg = b.topGainers()
print(tg)
```

Getting Top Losers

```python
tl = b.topLosers()
print(tl)
```

Getting Period Trend Data

Get historic price trend of a stock over certain fixed period of time. This data can be used to visualize the trends in the stock price.

The available period options are '1M', '3M', '6M' and '12M'

```python
his = b.getPeriodTrend('534976','6M')
print(his)
```

We got the tickers for S&P BSE SENSEX
Then we get the security codes which is tickers after removing .BO from it
Now we will get the Security ID of all the components of the  S&P BSE SENSEX stock

```python
bse_india = []

for i in tickers :
    bse_india.append(b.getQuote(i)['securityID'])
```

Print the security ID 

```python
print(bse_india)
```

Now add '.NS' to every security ID 

```python
data = [i + '.NS' for i in bse_india]
```

```python
data
```

Getting the historical data

```python
final_data = yt.download(data)
```

```python
final_data
```

Calculating Daily Returns

```python
returns = final_data['Adj Close'].pct_change()
returns
```

```python
# Removing NanN values
returns = returns.dropna()
returns
```

```python
# Cumulating the returns
(1 + returns).cumprod().plot(figsize =(10,10))
```
